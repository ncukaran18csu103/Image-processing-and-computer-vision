```python
r = np.arange(0, 256)
s = np.zeros(r.shape)
s[:] = 200
s[r<120] = 10
s[r>180] = 10
plt.subplot(1, 2, 1)
plt.plot(r, s)
plt.axis([0, 255, 0, 255])
plt.title('Binary thresholding function')

r = np.arange(0, 256)
s = np.arange(0, 256)
a = r>120
b = r<180
s[a * b] = 200
plt.subplot(1, 2, 2)
plt.plot(r, s)
plt.axis([0, 255, 0, 255])
plt.title('Gradual thresholding function')
```

```python
from PIL import Image #Python Imaging Library
A_gray = Image.open('dollar.jpg')
A_gray = A_gray.convert('L')

temp = np.asarray(A_gray.getdata(), dtype=np.float64).reshape((A_gray.size[1], A_gray.size[0]))
A_gr = np.asarray(temp, dtype=np.uint8)
```

```python
#BIT PLANE SLICING
```

```python
plane7 = A_gr & 128*np.ones(A_gr.shape).astype('uint8')
plane6 = A_gr &  64*np.ones(A_gr.shape).astype('uint8')
plane5 = A_gr &  32*np.ones(A_gr.shape).astype('uint8')
plane4 = A_gr &  16*np.ones(A_gr.shape).astype('uint8')
plane3 = A_gr &   8*np.ones(A_gr.shape).astype('uint8')
plane2 = A_gr &   4*np.ones(A_gr.shape).astype('uint8')
plane1 = A_gr &   2*np.ones(A_gr.shape).astype('uint8')
plane0 = A_gr &   1*np.ones(A_gr.shape).astype('uint8')

plt.figure(figsize=(20,7))
plt.subplot(2, 4, 1)
plt.imshow(plane7, cmap='gray')
plt.title('Plane 7 (MSB)')
plt.subplot(2, 4, 2)
plt.imshow(plane6, cmap='gray')
plt.title('Plane 6')
plt.subplot(2, 4, 3)
plt.imshow(plane5, cmap='gray')
plt.title('Plane 5')
plt.subplot(2, 4, 4)
plt.imshow(plane4, cmap='gray')
plt.title('Plane 4')
plt.subplot(2, 4, 5)
plt.imshow(plane3, cmap='gray')
plt.title('Plane 3')
plt.subplot(2, 4, 6)
plt.imshow(plane2, cmap='gray')
plt.title('Plane 2')
plt.subplot(2, 4, 7)
plt.imshow(plane1, cmap='gray')
plt.title('Plane 1')
plt.subplot(2, 4, 8)
plt.imshow(plane0, cmap='gray')
plt.title('Plane 0 (LSB)')
```

```python
plane7_6=cv2.add(plane7 ,plane6) 
plane7_6_5=cv2.add(plane5,plane7_6)
#plt.imshow(plane7_6,cmap='gray')
plt.imshow(plane7_6_5,cmap='gray')
```
